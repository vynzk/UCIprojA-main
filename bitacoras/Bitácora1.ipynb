## Tipos de señales

Aquí tienes un resumen de cada una de las señales comunes en el monitoreo cardiovascular y respiratorio:

### Señales del ECG (Electrocardiograma):

1. **Derivación II (ECG II)**:
   - Registra la diferencia de potencial eléctrico entre el **brazo derecho (RA)** y la **pierna izquierda (LL)**.
   - Es una de las derivaciones más comunes y útiles para observar la onda P y el ritmo cardíaco general.

2. **Derivación V (ECG V)**:
   - Es una **derivación precordial** que se coloca en el pecho (en posiciones específicas como V1, V2, etc.).
   - Se utiliza para observar la actividad eléctrica en diferentes partes del corazón, lo que es útil para detectar **infartos** y otras anomalías cardíacas.

3. **aVF (Augmented Vector Foot)**:
   - Es una derivación que mide la diferencia de potencial hacia el pie izquierdo en comparación con un punto de referencia entre el brazo derecho y el brazo izquierdo.
   - Ayuda a observar la actividad eléctrica en el plano frontal inferior del corazón.

4. **aVL (Augmented Vector Left)**:
   - Mide el potencial hacia el **brazo izquierdo** respecto a una combinación de electrodos en el brazo derecho y la pierna izquierda.
   - Se utiliza para evaluar el lado superior izquierdo del corazón.

5. **Derivación I (ECG I)**:
   - Mide la diferencia de potencial entre el **brazo derecho (RA)** y el **brazo izquierdo (LA)**.
   - Es útil para observar la actividad eléctrica a lo largo del plano horizontal del corazón.

6. **Derivación III (ECG III)**:
   - Mide la diferencia de potencial entre el **brazo izquierdo (LA)** y la **pierna izquierda (LL)**.
   - Complementa a las derivaciones I y II para un análisis más completo del corazón en el plano frontal.

7. **MCL (Mid-Clavicular Lead)**:
   - Es una derivación modificada de los precordiales (torácicas) que se coloca en una posición central, cerca de la línea media clavicular.
   - Se usa en situaciones específicas para una visión clara de la actividad eléctrica en el corazón.

### Otras Señales:

8. **PLETH (Fotopletismografía o Pletismografía de pulso)**:
   - Mide los cambios en el **volumen sanguíneo** en los tejidos, típicamente usando un sensor de luz.
   - Se utiliza en oxímetros de pulso para medir la **saturación de oxígeno (SpO2)** y el **ritmo cardíaco**.

9. **ABP (Presión Arterial Invasiva o Arterial Blood Pressure)**:
   - Mide la **presión arterial** de manera invasiva, directamente a través de un catéter en una arteria.
   - Es útil para monitorear de forma continua la presión arterial en pacientes críticos.

10. **RESP (Respiración)**:
    - Monitorea la frecuencia respiratoria midiendo los cambios en el volumen de aire en los pulmones o en el movimiento del tórax.
    - Se utiliza para evaluar la **frecuencia respiratoria** y la calidad de la respiración en situaciones de monitoreo clínico.

### Resumen:

| **Señal** | **Descripción** |
|-----------|-----------------|
| **ECG II** | Derivación entre el brazo derecho y la pierna izquierda; analiza el ritmo cardíaco general. |
| **ECG V**  | Derivación precordial en el pecho, útil para detectar infartos y otras anomalías. |
| **aVF**    | Derivación hacia el pie izquierdo, observa la parte inferior del corazón. |
| **aVL**    | Derivación hacia el brazo izquierdo, evalúa el lado superior izquierdo del corazón. |
| **ECG I**  | Derivación entre el brazo derecho e izquierdo; observa el plano horizontal del corazón. |
| **ECG III** | Derivación entre el brazo izquierdo y la pierna izquierda; complementa las derivaciones I y II. |
| **MCL**    | Derivación centralizada en el pecho, usada para ver la actividad cardíaca desde el centro del tórax. |
| **PLETH**  | Mide el volumen de sangre en los tejidos; se utiliza en oxímetros de pulso para SpO2 y ritmo cardíaco. |
| **ABP**    | Mide la presión arterial de manera invasiva. |
| **RESP**   | Monitorea la frecuencia respiratoria y la calidad de la respiración. |

Estas señales proporcionan una visión completa del estado cardiovascular y respiratorio de un paciente, permitiendo diagnósticos y monitoreos en tiempo real.



## Datos de .hea

### **Campos del Archivo `.hea`**

- **`record_name`**: `'a103l'`
  - El nombre del registro. Este es el identificador del conjunto de datos o registro de señales.

- **`n_sig`**: `3`
  - Número de señales en el registro. En este caso, hay tres señales.

- **`fs`**: `250`
  - Frecuencia de muestreo en Hz. Significa que las señales están muestreadas a 250 muestras por segundo.

- **`counter_freq`**: `None`
  - Frecuencia del contador, si se usa. No se proporciona en este caso.

- **`base_counter`**: `None`
  - Contador base, si se usa. No se proporciona en este caso.

- **`sig_len`**: `82500`
  - Longitud de la señal en muestras. Esto indica que cada señal tiene 82,500 muestras.

- **`base_time`**: `None`
  - Hora base del registro. No se proporciona en este caso.

- **`base_date`**: `None`
  - Fecha base del registro. No se proporciona en este caso.

- **`comments`**: `['Asystole', 'False alarm']`
  - Comentarios sobre el registro. En este caso, el registro está marcado como 'Asystole' y 'False alarm', indicando que es un evento de asistolia que ha sido clasificado como una alarma falsa.

- **`sig_name`**: `['II', 'V', 'PLETH']`
  - Nombres de las señales. Estas son las señales que se están registrando:
    - `'II'`: Derivación ECG II
    - `'V'`: Una señal adicional, probablemente una derivación de ECG o similar.
    - `'PLETH'`: Fotopletismograma, que mide el volumen de sangre en un punto del cuerpo.

- **`p_signal`**: `None`
  - Señales principales (p_signal). No está disponible en esta instancia, ya que los datos de señales se cargarán desde un archivo `.mat`.

- **`d_signal`**: `None`
  - Señales derivadas (d_signal). No está disponible en esta instancia.

- **`e_p_signal`**: `None`
  - Señales eléctricas (e_p_signal). No está disponible en esta instancia.

- **`e_d_signal`**: `None`
  - Señales derivadas eléctricas (e_d_signal). No está disponible en esta instancia.

- **`file_name`**: `['a103l.mat', 'a103l.mat', 'a103l.mat']`
  - Nombres de los archivos que contienen las señales. Aquí, todos los archivos de datos son `'a103l.mat'`.

- **`fmt`**: `['16', '16', '16']`
  - Formato de los datos en cada archivo. Los datos están en formato de 16 bits.

- **`samps_per_frame`**: `[1, 1, 1]`
  - Número de muestras por fotograma. Aquí cada señal tiene 1 muestra por fotograma.

- **`skew`**: `[None, None, None]`
  - Sesgo de lectura de datos. No está disponible en esta instancia.

- **`byte_offset`**: `[24, 24, 24]`
  - Desplazamiento en bytes para el inicio de cada señal dentro del archivo.

- **`adc_gain`**: `[7247.0, 10520.0, 12530.0]`
  - Ganancia del convertidor analógico-digital (ADC) para cada señal. Esto indica cómo se amplifican las señales antes de ser digitalizadas.

- **`baseline`**: `[0, 0, 0]`
  - Línea base de cada señal. Aquí todas las señales tienen una línea base de 0.

- **`units`**: `['mV', 'mV', 'NU']`
  - Unidades de las señales:
    - `'mV'`: Milivoltios para las señales ECG.
    - `'NU'`: Unidad no especificada para el fotopletismograma (posiblemente una unidad normalizada o sin unidades específicas).

- **`adc_res`**: `[16, 16, 16]`
  - Resolución del ADC en bits para cada señal. Aquí todas las señales tienen una resolución de 16 bits.

- **`adc_zero`**: `[0, 0, 0]`
  - Valor cero del ADC para cada señal. Aquí todos los valores son 0.

- **`init_value`**: `[-171, 9127, 6042]`
  - Valores iniciales en el archivo de datos para cada señal. Estos son los valores de referencia para el procesamiento de las señales.

- **`checksum`**: `[-27403, -301, -17391]`
  - Suma de comprobación de cada señal para verificar la integridad de los datos.

- **`block_size`**: `[0, 0, 0]`
  - Tamaño del bloque de datos para cada señal. Aquí el tamaño del bloque es 0.
