```python
# Funciones de obtencin de datos

def get_name_files(path,type):
    files = []
    for file in os.listdir(path):
        if file.endswith(type):
            files.append(file[:-4])
    return files

def read_file_hea(filename, path):
    # lee archivo .hea
    record = wfdb.rdheader(path+filename)
    return record

def read_file_mat(file, path):
    # carga archivo .mat (tiene las seales)
    mat_data = loadmat(path + file +'.mat')
    signals = mat_data['val']
    return signals

def read_file_flat(file,path):
    mat_data = loadmat(path + file +'.mat')
    if 'flatareas' in mat_data:
        flatareas = mat_data['flatareas']
    else:
        flatareas = np.array([])
    return flatareas

def read_file_zigzag(file,path):
    mat_data = loadmat(path + file +'.mat')
    if 'zigzagareas' in mat_data:
        zigzagareas = mat_data['zigzagareas']
    else:
        zigzagareas = np.array([])
    return zigzagareas
```

## 3.2.2 Limpieza de los datos

Gracias a las funciones de visualizacin se pueden observar caracteristicas importantes de la seal.

En primer lugar se observ estas no tenan una longitud estndar, muchas posean ruido y no se identificaba a simple vista el complejo QRS.
Adems muchos canales posean lneas planas y en zigzag, lo que dificulta an mas la deteccin de caractersticas.

Debido a esto se determina el siguiente plan de accin:
- Deteccin de lineas planas y zig_zag.
- Desarrollo de un filtro butterworth de paso bajo, con un cutoff=1 y order=2. [2]
- Clasificar datos vlidos para entrenamiento y testing.

Se detallan a continuacin funciones importantes:

**mark_flat_lines** : Marca lneas planas en la seal donde la diferencia entre valores consecutivos es menor que un umbral.

Parameteros:
- signal: La seal a analizar (ECG o PPG).
- fs: Frecuencia de muestreo en Hz.
- min_duration: Duracin mnima en segundos para que se considere una lnea plana.
- flat_value: El valor que se utilizar para marcar las lneas planas en la seal.
- threshold: Umbral de diferencia entre muestras consecutivas para considerar una lnea plana.

Returns:
- marked_signal: La seal con las lneas planas marcadas.
- flat_areas: Lista de tuplas (inicio, fin) que indican las reas planas detectadas.

**mark_zigzag_lines** :  Identifica segmentos en zigzag en una seal.
    
Parmetros:
- signal (numpy array): La seal a analizar.
- fs (int): Frecuencia de muestreo en Hz. Por defecto es 250 Hz.
- min_duration (float): Duracin mnima en segundos de un zigzag. Por defecto es 0.2 segundos.

Retorna:
zigzag_areas (list): Lista de tuplas (inicio, fin) de los segmentos en zigzag.

Tanto 'min_duration' como 'threshold' fueron ajustados para detectar correctamente seales sin actividad necesaria, o de nula actividad.

```python

def mark_flat_lines(signal, fs=250, min_duration=0.3, flat_value=-9999, threshold=2):
    # Nmero de muestras consecutivas que definen una lnea plana
    min_samples = int(min_duration * fs)

    # Diferencias absolutas entre muestras consecutivas
    diff_signal = np.abs(np.diff(signal))

    # Encuentra dnde la diferencia es menor que el umbral
    flat_segments = (diff_signal < threshold).astype(int)

    # Identifica segmentos planos continuos de al menos min_samples de duracin
    flat_areas = []
    current_length = 0
    start_index = None

    for i in range(len(flat_segments)):
        if flat_segments[i] == 1:  # Si es un segmento plano
            if current_length == 0:
                start_index = i
            current_length += 1
        else:
            if current_length >= min_samples:  # Si cumple con la duracin mnima
                flat_areas.append((start_index, start_index + current_length))
            current_length = 0

    # Considerar el ltimo segmento si termina en una zona plana
    if current_length >= min_samples:
        flat_areas.append((start_index, start_index + current_length))

    # Marca los segmentos planos en la seal
    marked_signal = np.copy(signal)
    for start, end in flat_areas:
        marked_signal[start:end] = flat_value  # Marcar con un valor especfico como -9999
    
    return marked_signal, flat_areas

def mark_zigzag_lines(signal, fs=250, min_duration=0.3):
    # Nmero de muestras consecutivas que definen un zigzag
    min_samples = int(min_duration * fs)
    
    # Calcula las diferencias entre muestras consecutivas (pendiente)
    diff_signal = np.diff(signal)
    
    # Determina si la pendiente es positiva o negativa
    sign_changes = np.sign(diff_signal)
    
    # Identifica los cambios de signo consecutivos (zigzag)
    zigzag_segments = (sign_changes[:-1] * sign_changes[1:] == -1).astype(int)

    # Identifica segmentos zigzag continuos de al menos min_samples de duracin
    zigzag_areas = []
    current_length = 0
    start_index = None
    
    for i in range(len(zigzag_segments)):
        if zigzag_segments[i] == 1:  # Cambio de pendiente
            if current_length == 0:
                start_index = i
            current_length += 1
        else:
            if current_length >= min_samples:
                zigzag_areas.append((start_index, start_index + current_length + 1))
            current_length = 0

    # ltimo segmento
    if current_length >= min_samples:
        zigzag_areas.append((start_index, start_index + current_length + 1))
    
    return zigzag_areas

# filtro butterworth
def butter_bandpass_filter(data, cutoff=1, fs=250, order=2):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y
```

Para la clasificacin de anomalas cardiacas se necesitan datos cuyas caracteristicas puedan ser identificables, debido a la naturaleza de los datos existen casos en donde ninguno de los tres o cuatro canales presentados son vlidos para extraccin de caractersticas, ya sea porque no poseen seales en las que se pueda identificar correctamente latidos cardiacos y complejos QRS o donde se presenten lneas planas o zig-zag que interrumpan, volviendo corrupto el canal.

Es por esto que se identifican y posteriormente se eliminan del set de entrenamiento.


```python

new_path = 'data/cleaned/'
print('Seales a limpiar:', nameFiles)

if not os.path.exists(new_path):
    os.makedirs(new_path)
    
#se separan las seales calificadas y las no calificadas
for fileName in nameFiles:
    path = old_path
    record = read_file_hea(fileName, path)
    signals = read_file_mat(fileName, path)

    # Recorta y deja slo los ltimos 10 segundos de seal
    signals = signals[:, -record.fs*20:]
    dirtySignals = signals

    # Se aplica filtro slo a las seales ECG ('I', 'II', 'III' y 'V')
    ecg_type = ['I', 'II', 'III', 'V']
    flat_areas_all = []
    zigzag_areas_all = []

    my_ecg = []
    strike = 0

    try:
        for i in range(len(record.sig_name)):
            if record.sig_name[i] in ecg_type and record.sig_name[i] not in my_ecg:
                my_ecg.append(record.sig_name[i])
                
        for i in range(len(record.sig_name)):

            if record.sig_name[i] in ecg_type:
                
                flat_signal, flat_areas0 = mark_flat_lines(signals[i, :], fs=record.fs)
                zigzag_areas0 = mark_zigzag_lines(signals[i, :], fs=record.fs)

            if len(flat_areas0) != 0 or len(zigzag_areas0) != 0:
                strike = strike + 1

        if strike >= len(my_ecg):
            raise Exception('Seal no califica para ser del training, no contiene ecg limpias'.format(fileName))

        else:
            savemat(new_path + fileName + '.mat', {
            'val': signals,
            })

            print('se guard la seal ', fileName, ' en el directorio cleaned')

    except Exception as e:
        print(f'Seal corrupta {fileName}: {e}')
        print(f'Seal no califica para ser del training {fileName} se descarta todo su archivo')

```

Luego se revisan cuantas seales quedaron calificadas para entrenar los modelos.

```python
# Se revisa cuntas seales se guardaron en el directorio cleaned
i=0
for file in os.listdir('data/cleaned/'):
    if file.endswith('.mat'):
        i += 1
print('cantidad de datos vlidos para training:', i)
```

### 3.2.2.2 Preparacin de los datos calificados


**Limpieza de los datos vlidos** 

Se utiliza la librera [Biosspy](https://biosppy.readthedocs.io/en/stable/) para normalizacin de los datos y extraccin de caractersticas, junto con un filtro de paso de banda. Posterior a esto se extraen las caracteristicas de cada caso y se guardan los canales ECG y PPG en un vector de canales.

Las caractersticas extradas corresponden a las relevantes en una seal ECG, sin embargo no todas fueron utilizadas.


```python

# Extraccin de caractersticas

# retorna los datos necesarios y limpios de cada caso
def biosppy_cases_data(new_path, old_path, nameFiles):

    data_cases = []
    
    for fileName in nameFiles:

        signals = read_file_mat(fileName, new_path)
        record = read_file_hea(fileName, old_path)

        ecg_type = ['II','V']
        pleth_type = ['PLETH']
        
        allSignals = []
        
        channels_type = []
        anomaly_type= record.comments[0]
        alarm_veracity = record.comments[1]

        min_heart_rate_pleth = 0
        max_heart_rate_pleth = 0
        min_heart_rate_ecg = 0
        max_heart_rate_ecg = 0

        prom_heart_rate_ecg = 0
        var_heart_rate_ecg = 0

        rpeaks_time_mean = 0
        rpeaks_time_std = 0
        rmssd = 0
        prop_diff_50ms = 0

        qrs_duration_mean = 0
        qrs_amplitude_mean = 0
        qrs_symmetry = 0

        II_mean = 0
        II_median = 0
        II_std = 0
        II_min = 0
        II_max = 0
        II_kurtosis = 0
        II_skewness = 0
        II_energy = 0

        V_mean = 0
        V_median = 0
        V_std = 0
        V_min = 0
        V_max = 0
        V_kurtosis = 0
        V_skewness = 0
        V_energy = 0

        PLETH_mean = 0
        PLETH_median = 0
        PLETH_std = 0
        PLETH_min = 0
        PLETH_max = 0
        PLETH_kurtosis = 0
        PLETH_skewness = 0
        PLETH_energy = 0

        
    

        for i in range(len(record.sig_name)): 
            ecg_if = (record.sig_name[i] in ecg_type)
            pleth_if = (record.sig_name[i] in pleth_type)
            channel_type = None
            if ecg_if or pleth_if:
                
                if ecg_if:
                    try:
                        channel = signals[i, :]
                        butterFilter = butter_bandpass_filter(channel, cutoff=1, fs=250, order=2)
                        channel = channel - butterFilter
                        
                        channel = ecg.ecg(signal=channel, sampling_rate=250, show=False)
                        channel_type = record.sig_name[i]

                        heart_rate_ecg = channel['heart_rate']
                        rpeaks = channel['rpeaks']
                        qrs = channel['templates']

                        if heart_rate_ecg.any() and min_heart_rate_ecg == 0:
                            min_heart_rate_ecg = min(heart_rate_ecg)
                        if heart_rate_ecg.any() and max_heart_rate_ecg == 0:
                            max_heart_rate_ecg = max(heart_rate_ecg)
                        if heart_rate_ecg.any() and prom_heart_rate_ecg == 0:
                            prom_heart_rate_ecg = np.mean(heart_rate_ecg)
                        if heart_rate_ecg.any() and var_heart_rate_ecg == 0:
                            var_heart_rate_ecg = np.var(heart_rate_ecg)

                        if rpeaks.any() and rpeaks_time_mean == 0:
                            rpeaks_time_mean = np.mean(np.diff(rpeaks))
                        if rpeaks.any() and rpeaks_time_std == 0:
                            rpeaks_time_std = np.std(np.diff(rpeaks))
                        if rpeaks.any() and rmssd == 0:
                            rmssd = np.sqrt(np.mean(np.diff(rpeaks)**2))
                        if rpeaks.any() and prop_diff_50ms == 0:
                            prop_diff_50ms = np.sum(np.diff(rpeaks) > 50) / len(rpeaks)

                        if qrs.any() and qrs_duration_mean == 0:
                            qrs_duration_mean = np.mean([len(qrs[i]) for i in range(len(qrs))])
                        if qrs.any() and qrs_amplitude_mean == 0:
                            qrs_amplitude_mean = np.mean([np.max(qrs[i]) for i in range(len(qrs))])
                        if qrs.any() and qrs_symmetry == 0:
                            qrs_symmetry = np.mean([np.max(qrs[i])/np.min(qrs[i]) for i in range(len(qrs))])

                        if record.sig_name[i] == 'II':
                            II_mean = np.mean(channel['filtered'])
                            II_median = np.median(channel['filtered'])
                            II_std = np.std(channel['filtered'])
                            II_min = np.min(channel['filtered'])
                            II_max = np.max(channel['filtered'])
                            II_kurtosis = scipy.stats.kurtosis(channel['filtered'])
                            II_skewness = scipy.stats.skew(channel['filtered'])
                            II_energy = np.sum(channel['filtered']**2)
                        elif record.sig_name[i] == 'V':
                            V_mean = np.mean(channel['filtered'])
                            V_median = np.median(channel['filtered'])
                            V_std = np.std(channel['filtered'])
                            V_min = np.min(channel['filtered'])
                            V_max = np.max(channel['filtered'])
                            V_kurtosis = scipy.stats.kurtosis(channel['filtered'])
                            V_skewness = scipy.stats.skew(channel['filtered'])
                            V_energy = np.sum(channel['filtered']**2)
                        

                        allSignals.append(channel)
                        if channel_type is not None:
                            channels_type.append(channel_type)

                    except Exception as e:
                        print(f'canal {record.sig_name[i]} en seal {fileName} no vlido para biosppy: {e}')
                    
                elif pleth_if:
                    
                    try:

                        channel = signals[i, :]
                        channel = ppg.ppg(signal=channel, sampling_rate=250, show=False)
                        
                        filtered_channel = channel['filtered']
                        channel_type = record.sig_name[i]

                        peaks = channel['peaks']
                        heart_rate_pleth = channel['heart_rate']

                        if heart_rate_pleth.any():
                            min_heart_rate_pleth = min(heart_rate_pleth)
                        if heart_rate_pleth.any():
                            max_heart_rate_pleth= max(heart_rate_pleth)

                        allSignals.append(channel)
                        if channel_type is not None:
                            channels_type.append(channel_type)
                        if peaks.any():
                            PLETH_mean = np.mean(filtered_channel)
                            PLETH_median = np.median(filtered_channel)
                            PLETH_std = np.std(filtered_channel)
                            PLETH_min = np.min(filtered_channel)
                            PLETH_max = np.max(filtered_channel)
                            PLETH_kurtosis = scipy.stats.kurtosis(filtered_channel)
                            PLETH_skewness = scipy.stats.skew(filtered_channel)
                            PLETH_energy = np.sum(filtered_channel**2)

                    except Exception as e:
                        print(f'canal {record.sig_name[i]} en seal {fileName} no vlido para biosppy: {e}')
        data_cases.append({
            'channels': allSignals, 
            'channels_type': channels_type,
            'alarm_veracity': alarm_veracity,
            'anomaly_type': anomaly_type,

            'max_heart_rate_ecg': max_heart_rate_ecg, 
            'min_heart_rate_ecg': min_heart_rate_ecg, 
            'max_heart_rate_pleth': max_heart_rate_pleth, 
            'min_heart_rate_pleth': min_heart_rate_pleth,

            'prom_heart_rate_ecg': prom_heart_rate_ecg,
            'var_heart_rate_ecg': var_heart_rate_ecg,
            'rpeaks_time_mean': rpeaks_time_mean,
            'rpeaks_time_std': rpeaks_time_std,
            'rmssd': rmssd,
            'prop_diff_50ms': prop_diff_50ms,
            'qrs_duration_mean': qrs_duration_mean,
            'qrs_amplitude_mean': qrs_amplitude_mean,
            'qrs_symmetry': qrs_symmetry,

            'II_mean': II_mean,
            'II_median': II_median,
            'II_std': II_std,
            'II_min': II_min,
            'II_max': II_max,
            'II_kurtosis': II_kurtosis,
            'II_skewness': II_skewness,
            'II_energy': II_energy,

            'V_mean': V_mean,
            'V_median': V_median,
            'V_std': V_std,
            'V_min': V_min,
            'V_max': V_max,
            'V_kurtosis': V_kurtosis,
            'V_skewness': V_skewness,
            'V_energy': V_energy,

            'PLETH_mean': PLETH_mean,
            'PLETH_median': PLETH_median,
            'PLETH_std': PLETH_std,
            'PLETH_min': PLETH_min,
            'PLETH_max': PLETH_max,
            'PLETH_kurtosis': PLETH_kurtosis,
            'PLETH_skewness': PLETH_skewness,
            'PLETH_energy': PLETH_energy,

            'II': 1 if 'II' in channels_type else 0,
            'V': 1 if 'V' in channels_type else 0,
            'PLETH': 1 if 'PLETH' in channels_type else 0
            
            })

    return data_cases
```

Claro, aqu tienes el resumen sin la palabra "relevancia":

**Frecuencia Cardaca Mxima del ECG (max_heart_rate_ecg)**
Indica episodios de taquicardia y permite evaluar la capacidad aerbica. Su medicin es crucial para prevenir riesgos en personas con enfermedades cardacas.

**Frecuencia Cardaca Mnima del ECG (min_heart_rate_ecg)**
Detecta bradicardia, lo que puede sealar bloqueos cardacos o problemas en la conduccin elctrica. Es esencial para la identificacin temprana de complicaciones.

**Raz Cuadrada de la Media de las Diferencias Cuadrticas Sucesivas (rmssd)**
Mide la variabilidad de la frecuencia cardaca. Alta variabilidad indica buena salud, mientras que baja puede asociarse a estrs o enfermedades cardacas.

**Tipo de Anomala (anomaly_type)**
Clasifica irregularidades en el ECG, como asistolia, ayudando en el diagnstico y tratamiento adecuado de problemas cardacos.

**Promedio de la Frecuencia Cardaca del ECG (prom_heart_rate_ecg)**
Proporciona una visin general del ritmo cardaco normal y ayuda a identificar desviaciones significativas.

**Varianza de la Frecuencia Cardaca del ECG (var_heart_rate_ecg)**
Refleja la estabilidad del ritmo cardaco. Cambios en la varianza pueden indicar estrs o condiciones patolgicas.

**Promedio del Tiempo entre Picos R (rpeaks_time_mean)**
Evala la consistencia en los intervalos entre latidos, til para identificar arritmias.

**Desviacin Estndar del Tiempo entre Picos R (rpeaks_time_std)**
Indica la regularidad del ritmo cardaco. Variaciones pueden sealar problemas elctricos en el corazn.

**Proporcin de Diferencias de Intervalo Mayor a 50 ms (prop_diff_50ms)**
Mide la variabilidad en los intervalos entre picos R, til para evaluar el riesgo de arritmias.

**Promedio de la Amplitud del QRS (qrs_amplitude_mean)**
Relacionada con la fuerza de contraccin ventricular; variaciones pueden indicar problemas de conduccin o isquemia.

**Simetra del Complejo QRS (qrs_symmetry)**
Proporciona informacin sobre la eficiencia de la despolarizacin ventricular; alteraciones pueden indicar problemas estructurales o funcionales.
